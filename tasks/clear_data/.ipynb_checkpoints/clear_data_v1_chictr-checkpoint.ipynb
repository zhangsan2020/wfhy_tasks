```python
import pymongo
import pandas as pd
```

```python
def get_sqldata(db,col):
    mongo_cli = pymongo.MongoClient('127.0.0.1',27017)
    col = mongo_cli[db][col]
    df = pd.DataFrame(list(col.find({'url_id':{'$gt':177308}})))
    return df
```

```python
df = get_sqldata('wfhy_update','chictr_all')
df.head()
```

### 根据手机号去重

## 分割手机号,得出11位数字的手机号码, 方便清洗, 否则部分 +86 xx 和  86 xx难以去重 

```python
df['phone'] = df.phone.map(lambda x: x.split(' ')[1] if len(x.split(' ')) > 1 else x)
```

```python
df.head()
```

## 判断真实手机号:  手机字符串包含为 (以 1 开始,其后有10位数字)

```python
new_df = df[df.phone.str.contains('1(\d+){10}')]
```

```python
new_df.head()
```

```python
new_df.info()
```

## 根据手机号去重,保留重复号码的第一个手机号

```python
new_df.drop_duplicates(subset='phone',inplace=True)
```

```python
new_df.info()
```

## 简单查阅是否还有重复

```python
new_df[new_df['email']=="2632494613@qq.com"]
```

```python
new_df[new_df.phone == '13596052383']
```

## 保存到excel中

## 摘取部分数据进行存储

```python
writer = pd.ExcelWriter(r'chictr_all_clear_20220915.xlsx', options={'strings_to_urls': False})
new_df.to_excel(writer,encoding='utf-8',index=False)
writer.save()
```
