### 根据手机号去重

## 分割手机号,得出11位数字的手机号码, 方便清洗, 否则部分 +86 xx 和  86 xx难以去重 

```python
df['phones'] = df.phones.map(lambda x: x.split(' ')[1] if len(x.split(' ')) > 1 else x)
```

```python
df.head()
```

## 判断真实手机号:  手机字符串包含为 (以 1 开始,其后有10位数字)

```python
new_df = df[df.phones.str.contains('1(\d+){10}')]
```
